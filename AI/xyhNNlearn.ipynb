# 首先，画出最终的输入输出函数3维曲线。

```python
画出它的输入输出函数3D图
提示：Z=numpy.logical.xor(X,Y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

x1=numpy.arange(0,2,1)
x2=numpy.arange(0,2,1)

X,Y=np.meshgrid(x1, x2)    # x-y 平面的网格
Z=numpy.logical_xor(X,Y)

ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = plt.get_cmap('rainbow'))
ax.contour(X, Y, Z, offset = -0.5, cmap = 'rainbow')
ax.set_zlim(-1, 1)

plt.show()

print(X)
print(Y)
print(Z)

```

# 定义NN输入输出函数

```python
import numpy as np

def NNWadd(x1,x2):
    
    X=x1,x2    
    
    print("x1")
    print(x1)
    print("x2")
    print(x2)
    
    w1=1
    w2=1
    w3=10
    w4=10
    w5=-10
    w6=10
    
    A=np.add(w1*x1,w2*x2)
    B=np.add(w3*x1,w4*x2)
    As=1/(1+np.exp(-A))
    Bs=1/(1+np.exp(-B))
    R=np.add(w5*As,w6*Bs)
    T=1/(1+np.exp(-R))
 
    
    print("T")
    print(T)
  
    return T
         

    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x1=np.arange(0,2,1)
x2=np.arange(0,2,1)

print("x1,x2")
print(x1)
print(x2)

X,Y=np.meshgrid(x1, x2) 

print("X,Y")
print(X)
print(Y)

NNWadd(X,Y)
```

# 画出NNWadd的三维图

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

x1=np.arange(0,2,1)
x2=np.arange(0,2,1)

X,Y=np.meshgrid(x1, x2)    # x-y 平面的网格
     
ZADD=NNWadd(X,Y)

ax.plot_surface(X, Y, ZADD, rstride = 1, cstride = 1, cmap = plt.get_cmap('rainbow'))
ax.contour(X, Y, ZADD, offset = -0.5, cmap = 'rainbow')
ax.set_zlim(-1, 1)

plt.show()

print(ZADD)
```
